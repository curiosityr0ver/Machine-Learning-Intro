```python
import numpy as np
import pandas as pd
```

```python
#jars sold
np.random.seed(0)
sales = np.random.randint(20, size = (5,3))
sales
```

```python
weeklySales = pd.DataFrame(sales, index = ["Mon", "Tue", "Wed", "Thurs", "Fri"], columns = ["Almond Butter", "Peanut Butter", "Cashew Butter"])
```

```python
weeklySales
```

```python
prices = np.array([10, 8, 15])
butter_prices = pd.DataFrame(np.reshape(prices, (1,3)) ,index=["Price"], columns = ["Almond Butter", "Peanut Butter", "Cashew Butter"])
butter_prices
```

```python
profit = np.dot(sales, prices).reshape(5,1)
```

```python
weekly_profit = pd.DataFrame(profit, index = ["Mon", "Tue", "Wed", "Thurs", "Fri"], columns = ["Profit"])
weekly_profit
```

```python
weeklySales["Revenue"] = weekly_profit
```

```python
weeklySales
```
