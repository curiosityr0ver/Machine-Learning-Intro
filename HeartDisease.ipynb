```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hds = pd.read_csv("13.1 heart-disease.csv")
fig, ax = plt.subplots()
ax.hist(hds["age"]);
ax.set(xlabel="Age", ylabel="Frequency");
```

###Formatting

```python
over50 = hds[hds["age"]> 50]
over50
```

```python
#PyPlot method
over50.plot(kind = "scatter", x = "age", y = "chol", c = "target");
```

```python
#OO Method
fix, ax = plt.subplots(figsize=(10, 6))
over50.plot(kind = "scatter", x = "age", y = "chol", c = "target", ax= ax)
ax.set(title = "Heart Disease and Cholestrol levels", xlabel="Cholestrol", ylabel="Prevalence")
ax.axhline(280, linestyle="--")
ax.set_xlim([50, 80]);
```

```python
#Subplot of age, target and thalach
```

```python
fig, ax = plt.subplots(nrows = 2, figsize = (10,10))
f1 = ax[0].scatter(x = over50["age"], y = over50["chol"], c = over50["target"])
f2 = ax[1].scatter(x = over50["age"], y = over50["thalach"], c = over50["target"])
ax[0].set(title="Heart Disease and Cholestrol levels", ylabel="Cholestrol")
ax[0].axhline(over50["chol"].mean(), linestyle= "--")
ax[0].legend(*f1.legend_elements(), title="Target", fancybox=True)
ax[1].set(title="Heart Disease and Heart Rate", xlabel="Age", ylabel = "Heart Rate");
ax[1].axhline(over50["thalach"].mean(), linestyle= "--")
ax[1].legend(*f2.legend_elements(), title="Target", fancybox=True)
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")
```
